```python
import pandas as pd
import numpy as np
from fuzzywuzzy import fuzz

df = pd.read_csv('/Users/mmorg27/Downloads/VOC/meta_Toys_and_Games.csv')
```

```python
display(df.head(5))
unique_names = df['brand'].unique()


for name in unique_names:
    if isinstance(name, str) and 'hasbro' in name.lower():
        df.loc[df['brand'].str.lower() == name.lower(), 'brand'] = 'Hasbro'

# Now 'hasbro' in any case in the 'brand' column has been replaced with 'Hasbro'


# Creating a new DataFrame containing rows where brand is "Hasbro"
hasbro_df = df[df['brand'].str.lower() == 'hasbro']

```

```python
from datetime import datetime

df2 = pd.read_csv('/Users/mmorg27/Downloads/VOC/Toys_and_Games.csv') 

# Convert 'reviewTime' column to datetime
df2['reviewTime'] = pd.to_datetime(df2['reviewTime'])

# Create a new 'rating' column with formatted values
df2['rating'] = df2['overall'].apply(lambda x: f"rating={x}")

# Display the modified DataFrame
print(df2.head())
```

```python
#display(hasbro_df)
df2.head(5)
# Extract the year and create a new 'year' column
df2['year'] = df2['reviewTime'].dt.year
merged_df = pd.merge(hasbro_df, df2, on='asin', how='inner')
display(merged_df)

merged_df.to_csv('hasbro_meta_reviews.csv', index=False)
```

```python
unique_titles = merged_df['title'].unique()

#for title in unique_titles:
   #print(title)

num_unique_titles = merged_df['title'].nunique()
#print("Number of unique titles:", num_unique_titles)

title_counts = merged_df['title'].value_counts().sort_values(ascending=True).head(10)
print(title_counts)

# Print all column names
print(merged_df.columns.tolist())

# Print unique observations of the 'overall' column
unique_overall = merged_df['overall'].unique()
print(unique_overall)
```

```python

# Group by 'title', calculate average 'overall', and sort in descending order
grouped_df = merged_df.groupby('title')['overall'].mean()

# Titles with rating averages greater than 4
high_rating_titles = grouped_df[grouped_df > 4].index.tolist()

# Titles with rating averages less than 3
low_rating_titles = grouped_df[grouped_df < 3].index.tolist()

# Create DataFrames for titles with rating averages greater than 4 and less than 3
high_rating_df = merged_df[merged_df['title'].isin(high_rating_titles)]
low_rating_df = merged_df[merged_df['title'].isin(low_rating_titles)]

# Print or use these DataFrames as needed
print("DataFrame for rating averages > 4:")
print(high_rating_df.head())

print("\nDataFrame for rating averages < 3:")
print(low_rating_df.head())

print(high_rating_df.describe())
print(low_rating_df.describe())

len(high_rating_df)
```

```python
competitors_df = not__hasbro_df
# Assuming hasbro_df and not_in_hasbro_df are your DataFrames
# Replace 'hasbro_df' and 'not_in_hasbro_df' with the names of your DataFrames

# Fill NaN values in 'title' columns with an empty string
hasbro_df['title'].fillna('', inplace=True)
competitors_df['title'].fillna('', inplace=True)

# Preprocess titles (lowercase and remove punctuation for example)
hasbro_titles_processed = merged_df['title'].str.lower().str.replace('[^\w\s]', '')
not_hasbro_titles_processed = competitors_df['title'].str.lower().str.replace('[^\w\s]', '')

# Define a threshold for similarity
threshold = 80  # Adjust as needed based on your data and similarity requirements

# Initialize a list to store matching titles
matching_titles = []

# Compare titles using fuzzy matching
for hasbro_title in hasbro_titles_processed:
    for not_hasbro_title in not_hasbro_titles_processed:
        similarity_score = fuzz.ratio(hasbro_title, not_hasbro_title)
        if similarity_score >= threshold:
            matching_titles.append((hasbro_title, not_hasbro_title, similarity_score))

# Display matching titles and their similarity scores
for match in matching_titles:
    print(f"Hasbro Title: {match[0]} | Non-Hasbro Title: {match[1]} | Similarity Score: {match[2]}")
```
